```python
df_dados = pd.read_csv('gastos-deputados.csv', sep=';')
```

```python
df_dados.head()
```

```python
qt_candidatos_uf = df_dados[['sgUF','ideCadastro']].groupby('sgUF', as_index=False).count()
qt_candidatos_uf
```

```python
!pip install altair
```

```python
import altair as alt
```

```python
alt.Chart(qt_candidatos_uf, title='titulo bla bla').mark_bar().encode(
    x=alt.X('sgUF', title='Estado'),
    y=alt.Y('ideCadastro', title='candidatos por estado'),
    color= alt.Color('sgUF')
)
```

```python
qt_candidatos_partido = df_dados[['sgPartido','ideCadastro']].groupby('sgPartido', as_index=False).count()
qt_candidatos_partido
```

```python
alt.Chart(qt_candidatos_partido, title='titulo bla bla2').mark_bar().encode(
    x=alt.X('sgPartido', title='Partido'),
    y=alt.Y('ideCadastro', title='candidatos por partido'),
    color= alt.Color('sgPartido')
)
```
